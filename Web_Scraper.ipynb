## Query Creation

```
import pandas as pd
init = {"JP Morgan Chase Bank","Bank of America Corporation","Citigroup","WELLS FARGO BANK, N.A.",
        "Goldman Sachs & Co. LLC","Morgan Stanley & Co. LLC","U.S. Bancorp","Truist Financial",
        "The PNC Financial Services Group, Inc.","TD Bank, NA","CAPITAL ONE SERVICES, LLC",
        "The Bank of New York Mellon","TIAA","Charles Schwab & Co., Inc.","HSBC Bank USA","State Street Corporation",
        "AMERICAN EXPRESS TRAVEL RELATED SERVICES COMPANY, INC.","Ally Financial","MUFG Union Bank, N.A.","Fifth Third Bank",
        "State Farm Mutual Automobile Insurance Company","USAA","Citizens Financial Group","BMO Harris Bank",
        "Barclays","KeyCorp","UBS","Santander Bank","Deutsche Bank","Regions Financial Corporation"}
quer = []
q2 = []
for i in init:
  for year in range(2000,2020):
    q2.append(i)
    x = '(AN/"'+ str(i) + '\") AND (APD/1/1/' + str(year) + "->12/31/"+str(year) +")"
    quer.append(x)

query = pd.DataFrame(q2,columns=['BankNames'])
query['QueryAPD'] = quer
query
```

```
queryISD = []
for i in init:
  for year in range(2000,2020):
    x = '(AN/"'+ str(i) + '\") AND (ISD/1/1/' + str(year) + "->12/31/"+str(year) +")"
    queryISD.append(x)

query['QueryISD'] = queryISD
query
```

## Convert Query to URL

```
common_url = "http://patft.uspto.gov/netacgi/nph-Parser?Sect1=PTO2&Sect2=HITOFF&u=%2Fnetahtml%2FPTO%2Fsearch-adv.htm&r=0&p=1&f=S&l=50&Query=%28AN%2F%22"
url_list = []

for i in init:
  for year in range(2000,2020): 
    temp = i.split(sep = ' ')
    url = common_url

    for words in temp[:-1]:
      if(words[-1] == ','):
        url += words[:-1]+'%2C+'
      else:
        url += words+'+'
    url += temp[-1]

    url += '%22%29+AND+%28ISD%2F1%2F1%2F' + str(year) + '-%3E12%2F31%2F' + str(year) + '%29&d=PTXT'
    
    # print(url) 
    url_list.append(url)
print(len(url_list))
query['URL_ISD'] = url_list
```
